```python
merged = None
for subdir, purpose in zip(subdirs, purposes):
    testdf = gp.read_file(basedir / subdir)
    testdf['label'] = subdir.split('/')[0]
    testdf['year'] = int(subdir[:4])
    testdf['month'] = int(subdir[4:6])
    testdf['total lines'] = len(testdf)
    testdf['kind'] = purpose
#    print(testdf.head(3))
    if merged is not None:
        merged = merged.append(testdf)
    else:
        merged = testdf
```

```python
merged.rename(columns={'FID': 'FlightlineID'})
merged.reset_index(drop=True, inplace=True)
```
